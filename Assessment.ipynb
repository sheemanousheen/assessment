```python
dict1 = {1: ["Male", 171, 96],
     2: ["Male", 161, 85],
     3: ["Male", 180, 77],
     4: ["Female",166,62],
     5: ["Female",150,70],
     6:["Female",167,82]
     }
  
# Print the names of the columns.
print ("{:<10} {:<10} {:<10}".format('Gender', 'Height', 'Weight'))
  
# print each data item.
for key, value in dict1.items():
    Gender, Height, Weight= value
    print ("{:<10} {:<10} {:<10}".format(Gender, Height, Weight ))

weight = eval(input('Your weight in kilos: ')) # <=== 75
height = eval(input('Your height in meters: ')) # <=== 1.75
def body_mass_index():
  bmi = weight / (height/100) ** 2
  print(f"You BMI is {bmi}")
  if bmi <= 18.4:
    print('Underweight.')
    print("you are at Malnutrition risk")
  elif 18.5 < bmi <= 24.9:
    print('Normal weight.')
    print("you are at Low risk")
  elif 25 < bmi <= 29.9:
    print('Overweight.')
    print("you are at Enhanced risk")
  elif 30 < bmi <=34.9:
    print('Moderately obese.')
    print("you are at Medium risk")
  elif 35 < bmi <=39.9:
    print('Severely obese.')
    print("you are at High risk")
  else:
    print('Very severely obese.')
    print("you are at Very high risk")
body_mass_index() 
```
